```python
import random

# Define possible actions
actions = ["move_left", "move_right", "move_up", "move_down", "stay"]

class RandomReflexAgent:
    def __init__(self): # Fix typo here
        pass

    def perceive(self, environment):
        # The agent perceives the current state of the environment
        # In this case, it's just a placeholder function
        return environment.get_state()

    def decide(self, perception):
        # Randomly choose an action based on the current perception
        return random.choice(actions)

    def act(self, action):
        print(f"Agent is performing action: {action}")

# Example environment (you can modify this)
class Environment:
    def __init__(self): # Fix typo here
        self.state = "normal"

    def get_state(self):
        # Return a random state for the environment
        return random.choice(["normal", "danger", "goal"])

# Simulate the Random Reflex Agent in an environment
if __name__ == "__main__": # Fix typo here
    env = Environment()
    agent = RandomReflexAgent()

    for step in range(10):  # Simulate for 10 steps
        print(f"\nStep {step + 1}")
        perception = agent.perceive(env)
        print(f"Agent perceives: {perception}")
        action = agent.decide(perception)
        agent.act(action)
```
