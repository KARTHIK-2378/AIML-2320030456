```python
from collections import deque
def bfs(g,s):
  q = deque([s])
  v = set()
  v.add(s)
  while q:
    n = q.popleft()
    print(n,end=" ")
    for i in g[n]:
      if i not in v:
        v.add(i)
        q.append(i)

g={'A':['B','C'],
   'B':['D','E'],
   'C':['F'],
   'D':[],
   'E':[],
   'F':[]
   }
bfs(g,'A')
```

```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v):
        if u not in self.graph:
            self.graph[u] = []
        self.graph[u].append(v)

    def dfs(self, start):
        visited = set()

        def dfs_util(node):
            visited.add(node)
            print(node)

            for neighbor in self.graph.get(node, []):
                if neighbor not in visited:
                    dfs_util(neighbor)

        dfs_util(start)
g = Graph()
g.add_edge(0, 1)
g.add_edge(0, 2)
g.add_edge(1, 2)
g.add_edge(2, 0)
g.add_edge(2, 3)
g.add_edge(3, 3)

print("Depth First Traversal:")
g.dfs(2)
```

```python
def iterative_deepening_search(node, goal, depth_limit):
    for depth in range(depth_limit):
        result = depth_limited_search(node, goal, depth)
        if result == 'found':
            return 'Goal found at depth: {}'.format(depth)
    return 'Goal not found within depth limit'

def depth_limited_search(node, goal, depth):
    if node == goal:
        return 'found'
    if depth == 0:
        return 'cutoff'
    cutoff_occurred = False
    for child in expand(node):
        result = depth_limited_search(child, goal, depth - 1)
        if result == 'found':
            return 'found'
        if result == 'cutoff':
            cutoff_occurred = True
    if cutoff_occurred:
        return 'cutoff'
    return 'not found'

def expand(node):
    children = {
        'A': ['B', 'C'],
        'B': ['D', 'E'],
        'C': ['F', 'G'],
        'D': [],
        'E': [],
        'F': [],
        'G': []
    }
    return children.get(node, [])
initial_node = 'A'
goal_node = 'D'
depth_limit = 3
print(iterative_deepening_search(initial_node, goal_node, depth_limit))
```
